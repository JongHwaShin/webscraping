```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.daum.net/'
res = requests.get(url)
print(res.ok)
```

```python
if res.ok:
    html = res.text
    soup = BeautifulSoup(html,'html.parser')
strong_tag_list = soup.select("strong[class='tit_g']") ### select("strong.tit_g") 도 가능하다 .은 class 찾는거 #은 id
# print(a_tag_list)
f = open('뉴스.txt','w')
for strong_tag in strong_tag_list:
    a_tag = strong_tag.select_one("a") #select_one("tag이름") 찾고싶은 tag이름 찾는거 -> 처음나온 태그 1개만 찾는다!
                                       #seledt("tag이름") 찾고싶은 tag이름 찾는거 -> 찾는 모든 tag찾음!
    print("링크: " + a_tag.get('href'))
    print("제목: " + strong_tag.text.strip()) ## .text 와 .string 둘다 사용가능!
    print('-----------------------------------------')
    f.write("링크: "+a_tag.get('href'))
    f.write('\n')
    f.write("제목: "+strong_tag.text.strip())
    f.write('\n')

f.close()


```
